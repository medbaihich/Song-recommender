```python
import turicreate as tc
```

```python
song = tc.SFrame('song_data.sframe')
```

```python
song.head()
```

```python
users= song['user_id'].unique()
```

```python
len(users)
```

```python
train_data, test_data = song.random_split(.8,seed=0)
```

```python
popularity_model=tc.popularity_recommender.create(train_data, user_id='user_id',item_id='song')
```

```python
popularity_model.recommend(users=[users[0]])
```

```python
popularity_model.recommend(users=[users[1]])
```

```python
personalized_model=tc.item_similarity_recommender.create(train_data,user_id='user_id',item_id='song')
```

```python
personalized_model.recommend(users=[users[0]])
```

```python
personalized_model.recommend(users=[users[1]])
```

```python
personalized_model.get_similar_items(['With Or Without You - U2','Chan Chan (Live) - Buena Vista Social Club'])
```

```python
model_performence= tc.recommender.util.compare_models(test_data,[popularity_model,personalized_model],user_sample=0.05)
```
